```python
# 导入 PyTorch 库
import torch
import torchvision

# 1. 检查 PyTorch 版本（确认安装版本）
print("PyTorch 版本：", torch.__version__)
print("TorchVision 版本：", torchvision.__version__)

# 2. 检查 CUDA/Metal 支持（适配你的 Mac 环境）
# Apple Silicon (M1/M2/M3) 芯片检查 Metal 加速
if torch.backends.mps.is_available():
    print("✅ Metal 加速可用（M 系列芯片）")
    # 测试张量运算
    x = torch.tensor([1.0, 2.0, 3.0], device="mps")
    print("Metal 张量运算测试：", x * 2)
# Intel 芯片检查 CPU 运行（无 CUDA）
elif not torch.cuda.is_available():
    print("✅ CPU 环境正常（无 GPU 加速）")
    # 测试基础张量运算
    x = torch.tensor([1.0, 2.0, 3.0])
    print("CPU 张量运算测试：", x * 2)

# 3. 检查 torchvision 基础功能
from torchvision import datasets, transforms
print("✅ torchvision 导入成功，基础功能可用")
```
